```python
import pandas as pd
from scipy.stats import skew, kurtosis

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

skewness = df[numeric_columns].apply(lambda x: skew(x.dropna()))
kurtosis_vals = df[numeric_columns].apply(lambda x: kurtosis(x.dropna()))

stats_df = pd.DataFrame({
    'Skewness': skewness,
    'Kurtosis': kurtosis_vals
})

print("Skewness and Kurtosis of Numeric Variables:")
print(stats_df)

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
descriptive_stats = df.describe()
print(descriptive_stats)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
plt.figure(figsize=(5,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

```python
def plot_bar_activity_level(df):
    """
    Bar chart comparing Physical Activity Level across Sleep Disorders.
    """
    plt.figure(figsize=(5, 4))
    sns.countplot(data=df, x='Physical Activity Level', hue='Sleep Disorder', palette='Blues')
    plt.title('Physical Activity Level across Sleep Disorders')
    plt.xlabel('Physical Activity Level')
    plt.ylabel('Count')
    plt.legend(title='Sleep Disorder')
    plt.show()
plot_bar_activity_level(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_sleep_duration_vs_age(df):
    """
    Line graph showing Sleep Duration vs. Age without grouping.
    """
    plt.figure(figsize=(5, 4))
    sns.lineplot(data=df, x='Age', y='Sleep Duration', marker='o', linestyle='-')
    plt.title('Sleep Duration vs. Age')
    plt.xlabel('Age')
    plt.ylabel('Sleep Duration (hours)')
    plt.grid()
    plt.show()
plot_sleep_duration_vs_age(df)
```

```python
def plot_box_sleep_bmi_with_hue(df):
    """
    Box plot showing the distribution of Sleep Duration across BMI Categories, differentiated by Sleep Disorder.
    """
    plt.figure(figsize=(5, 4))
    sns.boxplot(data=df, x='BMI Category', y='Sleep Duration', hue='Sleep Disorder', dodge=True , palette='Blues')
    plt.title('Sleep Duration by BMI Category')
    plt.xlabel('BMI Category')
    plt.ylabel('Sleep Duration (hours)')
    plt.legend(title='Sleep Disorder') 
    plt.show()
plot_box_sleep_bmi_with_hue(df)

```

```python
def plot_scatter_sleep_duration(df):
    """
    Scatter plot of Sleep Duration vs. Age, differentiated by Gender.
    """
    plt.figure(figsize=(6, 5))
    sns.scatterplot(data=df, x='Age', y='Sleep Duration', hue='Gender', palette='Blues', alpha=0.7)
    plt.title('Sleep Duration vs. Age by Gender')
    plt.xlabel('Age')
    plt.ylabel('Sleep Duration (hours)')
    plt.legend(title='Gender')
    plt.show()

plot_scatter_sleep_duration(df)
```
